# Welcome to RNN and LSTM: Learning Sequences Like a Storyteller!

Hello everyone! Today, we're diving into Recurrent Neural Networks (RNN) and Long Short-Term Memory (LSTM) networks. Think of this as learning how computers can understand stories, conversations, or any sequence of things that happen one after another. We'll use simple analogies from everyday life—no math formulas, just fun examples. Let's make this like chatting over coffee!

By the end of this session, you'll see how these models predict the next word in a sentence, just like your phone's keyboard suggests words while typing.

## What is Sequential Data?

Imagine you're telling a story to a friend: "Once upon a time, there was a brave knight who fought a dragon." Each word comes one after another, and the meaning builds up step by step.

Sequential data is just like that—information that has an order and depends on what came before. Here are some real-life examples:

- **Sentences and Stories**: In a book, words form sentences. If you shuffle them, the story makes no sense! Like: "Knight dragon fought brave a the" – huh?
  
- **Daily Conversations**: When you chat with someone: "How are you?" → "I'm good, thanks!" The response depends on the question before it.

- **Other Examples**: Music notes in a song (one note leads to the next), stock prices over days (today's price relates to yesterday's), or even steps in a recipe (mix flour first, then add eggs).

Why care? In real life, most things aren't standalone—they're connected in a sequence. Computers need to handle this to predict weather, suggest movies, or complete your emails.

Visualize it: Picture a chain of paper clips. Each clip links to the previous one. Break the chain, and it falls apart!

---

## Why Traditional Neural Networks Fail for Sequences

Traditional neural networks are like a snapshot camera—they look at one picture at a time but forget the whole album.

Real-life analogy: Imagine reading a long sentence: "The cat that chased the mouse that ate the cheese that was on the table finally got tired." If you forget the early words ("the cat") by the end, you lose the meaning!

Traditional networks treat each input independently. They don't remember what happened before, so for sequences:

- They "forget" earlier parts, like amnesia in a conversation.
- Can't capture dependencies: In "I grew up in France... I speak fluent _____," they might not link "France" to "French."

This is why we need something smarter that "remembers" as it goes along.

---

## What is RNN (Recurrent Neural Network)?

RNN is like a person reading a book word by word, remembering key parts to understand the plot.




- **How it Works**: It processes data one step at a time, like you reading: Word 1 → Word 2 → Word 3. At each step, it uses info from the previous step.

- **Hidden State as “Short-Term Memory”**: Think of the hidden state as your brain's short-term memory. It holds a summary of what you've read so far. For example:
  - Reading "The quick brown...": Memory notes "Something fast and animal-like."
  - Then "fox": Memory updates to "A fast brown fox."

RNNs are great for short stories but struggle with epics!




---

## Limitations of RNN

RNNs are like a storyteller who starts strong but forgets the beginning by the end.

- **Long-Term Dependency Problem**: Analogy: In a long fairy tale, "Once upon a time, a princess lost her glass slipper... [many pages later]... and the prince found it." If the storyteller forgets the "glass slipper" early on, the ending doesn't connect!

- **Forgetting Important Information**: Over long sequences, the "memory" fades. Early info gets diluted, like whispering a message in a long chain of people—it gets distorted.

This is called the "vanishing gradient" issue, but think of it as memory overload: Too much info, and the important bits get lost.

---

## Why LSTM Was Introduced

LSTM is like a smarter reader with a notebook and highlighter— it decides what to remember, forget, or use right now.

- Introduced to fix RNN's forgetting problem.
- LSTM has a "memory system" that handles long stories better, like a detective noting clues without losing track.

Emphasis: It's all about smart decision-making for memory—keep what's useful, ditch the rest, and use what's needed now. No more amnesia!



---

## LSTM Cell Explained Using Real-Life Analogies

An LSTM cell is like a librarian managing books in a library (the memory).

- **Forget Gate**: Like deciding to toss out old, irrelevant books. "Do I need this detail about the weather? No? Forget it!" Keeps memory clean.






- **Input Gate**: Decides what new info to add to the shelves. "This new clue about the character is important—add it!" It chooses what's worth remembering.







- **Output Gate**: Decides what info to pull out and use now. "For this part of the story, I need the hero's name—output that!"

Together, it's like your brain during a movie: Forget distractions, remember plot twists, and recall details when needed.





---

## Layers in an LSTM-Based Model

Think of the model as a team working on a puzzle:

- **Embedding Layer (Word Meaning Mapper)**: Turns words into "meanings." Like a translator: "Apple" could mean fruit or company— it maps to a code that captures context.

- **LSTM Layer (Memory + Sequence Understanding)**: The core team member who remembers the sequence and makes sense of it, using the gates we talked about.

- **Dense Layer (Final Decision Maker)**: The boss who looks at everything and decides the output, like picking the next word.

Stacked together: Input words → Map meanings → Process sequence with memory → Decide output.

## What is Next Word Prediction?

It's like your phone guessing the next word as you type: "Happy birthday to _____ " → "you"!

- **Real-Life Examples**:
  - Mobile keyboards: Type "Let's meet at" → Suggests "the park."
  - Gmail: "I'll send the report by" → "tomorrow."
  - Google Search: "How to bake" → "a cake."

- **Simple Sentence Completion**: "The sun rises in the _____ " → "east." Models learn from tons of text to predict logically.

Fun fact: It's sequence learning in action—understanding patterns to guess next.

---

## Epochs

Epochs are like revision cycles before an exam—you go through your notes multiple times to learn better.

- Each epoch: The model "reads" the entire dataset once, learning a bit more each time.
- Analogy: First read: Skim and get basics. Second: Dive deeper. Too many? Overkill and tiring.

Simple Guideline: Start with 5-10 epochs for small datasets. Watch if it improves—stop if it stops learning!

---

## Learning Curves

These are graphs showing how the model improves, like a fitness tracker for learning.

- **Loss Curve**: "Learning improvement over time." High loss at start (many mistakes), decreases as it learns (fewer errors). Like your exam scores improving with practice.

Visual: A line graph starting high and dipping down—flattening means it's learned enough.

---

## Evaluation Metrics (Simple & Practical)

Ways to check if the model is doing well, like grading a student's essay.

- **Accuracy**: How often it predicts correctly. Use for simple tasks, but not always best for words (many possible right answers).

- **Loss**: Measures "how wrong" predictions are. Lower is better—focus here, as it shows overall improvement.

- **Perplexity (Optional)**: Intuitive: How "surprised" the model is by real text. Low perplexity = good predictions, like not being shocked in a predictable story.

---

## Model Training Steps (End-to-End Flow)

Like baking a cake: Gather ingredients → Mix → Bake → Taste → Adjust.

- **Data**: Collect text (stories, sentences).
- **Model**: Build the LSTM "recipe."
- **Training**: Feed data, adjust over epochs.
- **Prediction**: Ask for next word.
- **Evaluation**: Check loss/accuracy.

Full flow: Prep data → Build model → Train → Predict → Evaluate results.



---

## Final Model Interpretation

How to judge if it's learning: Look at loss decreasing (good sign!). Predictions should make sense, like completing sentences logically.

If loss stays high: Maybe needs more epochs or better data. If predictions are weird: Overfitting (memorized too specifically) or underfitting (didn't learn enough).

Think: Is it like a good storyteller? If yes, success!

```python
# STEP 0: Imports
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
import matplotlib.pyplot as plt
```

```python
# STEP 1: Raw Text  (your 50-line corpus)

text = """
The cat sat on the mat.
The dog chased the cat.
Birds fly in the sky.
The sun shines bright.
Rain falls from the clouds.
Kids play in the park.
The baby laughs loudly.
The boy kicks the ball.
The girl reads a book.
The teacher writes on the board.
The bird sings on the tree.
The fish swims in the pond.
The flowers bloom in the garden.
The leaves fall from the tree.
The wind blows softly.
The stars shine at night.
The moon glows in the sky.
The car stops at the signal.
The bus arrives at the station.
The train moves on the track.
The clock ticks on the wall.
The child drinks milk.
The mother cooks the food.
The father drives the car.
The dog sleeps on the floor.
The cat drinks water.
The boy draws a picture.
The girl sings a song.
The rain makes the road wet.
The sun dries the clothes.
The kids jump in the puddles.
Birds sit on the wires.
The airplane flies in the sky.
The clouds cover the sun.
The ice melts in the heat.
The baby sleeps in the crib.
The phone rings on the table.
The door opens slowly.
The light turns on.
The fan spins fast.
The dog barks at the stranger.
The cat hides under the bed.
The boy runs in the field.
The girl dances on the stage.
The child rides a bicycle.
The friends play in the ground.
The teacher asks a question.
The student answers softly.
The family eats dinner together.
The day ends with a quiet night.
"""

corpus = [line.strip() for line in text.strip().split("\n") if line.strip()]
print("Number of lines in corpus:", len(corpus))
```

```python
# Step 2: Word → Number dictionary(Text Preprocessing)
# Tokenize: Turn words into unique numbers
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
word_index = tokenizer.word_index  # Dictionary of words to numbers
total_words = len(word_index) + 1  # +1 for padding
print("\nWord Index:", word_index)
```

```python
# STEP 3: Create sequences for next-word prediction

input_sequences = []

for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]
    # create n-gram sequences: [w1,w2]→w3, [w1,w2,w3]→w4, ...
    for i in range(2, len(token_list) + 1):
        n_gram_seq = token_list[:i]
        input_sequences.append(n_gram_seq)

print("Number of sequences:", len(input_sequences))
print("Example sequence (token ids):", input_sequences[0])
```

```python
# Show original sentences and their token lists
for i, line in enumerate(corpus[:3]):  # First 3 sentences
    tokens = tokenizer.texts_to_sequences([line])[0]
    print(f"Line {i+1}: '{line}' → tokens: {tokens}")

    # Show n-gram sequences created from this line
    print("  Created sequences:")
    for j in range(2, len(tokens) + 1):
        seq = tokens[:j]
        print(f"    {seq} → predict {tokens[j-1]}")
    print()
```

```python
# STEP 4: Padding (fixed size) and split into X, y

max_seq_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_seq_len, padding='pre')

X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# one-hot encode labels
y = tf.keras.utils.to_categorical(y, num_classes=total_words)

print("Max sequence length:", max_seq_len)
print("X shape:", X.shape)
print("y shape:", y.shape)
```

 Example:
X = numbers for "The cat sat on the"
y = 1 at position of "mat" (and 0 everywhere else)


```python
# STEP 5: Model Build (Embedding + LSTM + Dense)

embedding_dim = 16
lstm_units = 64

model = Sequential([
    Embedding(input_dim=total_words, output_dim=embedding_dim, input_length=X.shape[1]),
    LSTM(lstm_units),
    Dense(total_words, activation='softmax')
])

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.summary()
```

**Embedding**→ like giving each word a secret personality (10 numbers describe it)

1. input_dim=total_words (e.g., 87)

Vocabulary size = total unique words + 1

Creates a lookup table with 87 rows (one row per word)

Example: word "cat" = index 5 → looks up row #5


2. output_dim=embedding_dim (e.g., 16)

Vector size for each word

Each word becomes a 16-number vector

"cat" → [0.2, -0.1, 0.8, ...] (16 numbers)


3. input_length=X.shape[1] (e.g., 8)

Sequence length = how many words in each input

Every training example has exactly 8 words

Input shape: (batch_size, 8)
**LSTM(50)** → has 50 little memory workers remembering patterns

**Dense** → the final voter who says: “I think the next word is ‘mat’ with 70% confidence



```python
# STEP 6: Train (epochs)  [web:11][web:60]

epochs = 100
batch_size = 16

history = model.fit(
    X, y,
    epochs=epochs,
    batch_size=batch_size,
    verbose=1
)
```

```python
# STEP 7: Predict – next word helper

def predict_next_word(seed_text, tokenizer, max_seq_len, model):
    token_list = tokenizer.texts_to_sequences([seed_text])[0]
    token_list = pad_sequences([token_list], maxlen=max_seq_len - 1, padding='pre')
    predicted_probs = model.predict(token_list, verbose=0)
    predicted_index = np.argmax(predicted_probs, axis=-1)[0]
    for word, index in tokenizer.word_index.items():
        if index == predicted_index:
            return word
    return None
```

```python
# STEP 7 (continued): Try some predictions

test_inputs = [
    "The cat",
    "The dog",
    "The sun",
    "Kids play",
    "The child"
]

for s in test_inputs:
    next_w = predict_next_word(s, tokenizer, max_seq_len, model)
    print(f"Input: '{s}' → Predicted next word: '{next_w}'")
```

```python
# STEP 8 & 9: Loss and accuracy + curves

plt.figure(figsize=(12, 4))

# Accuracy curve
plt.subplot(1, 2, 1)
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training Accuracy over Epochs")
plt.legend()

# Loss curve
plt.subplot(1, 2, 2)
plt.plot(history.history["loss"], label="Training Loss", color="orange")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training Loss over Epochs")
plt.legend()

plt.tight_layout()
plt.show()
```
